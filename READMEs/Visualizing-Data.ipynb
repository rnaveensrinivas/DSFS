# Visualizing Data

Visualization is a fundamental skill required for practicing data science! There are two primary uses for data visualization: 
* **To Explore Data**: to understand the structure and gain insights
* **To Communicate Data**: to share the insights gained

> Note: It's easy to create visualizations, but to create impactful and readable ones is difficult. 

# Matplotlib

Matplotlib is one of the most widely used libraries in Python for **data visualization**. It allows you to create static, interactive, and animated plots to visualize data and understand trends, patterns, and distributions. Matplotlib is highly customizable, making it suitable for creating everything from simple plots to complex visualizations.

---

### **Key Features of Matplotlib**
1. **Versatile Plotting Capabilities**:
   - Line plots, scatter plots, bar plots, histograms, pie charts, and more.
   - Customizable styles, colors, and markers.

2. **Fine-grained Control**:
   - Customize axes, labels, legends, titles, and annotations.
   - Adjust figure size, resolution, and layout.

3. **Integration with Other Libraries**:
   - Works seamlessly with libraries like NumPy, pandas, and seaborn.
   - Outputs plots into multiple formats (PNG, PDF, SVG, etc.).

4. **Interactive Features**:
   - Interactive navigation (zooming, panning) in graphical user interfaces (GUIs).

---

### **Commonly Used Plot Types**

| Plot Type           | Method                | Description                               |
|---------------------|-----------------------|-------------------------------------------|
| **Line Plot**       | `plt.plot()`          | Connects points with lines.               |
| **Scatter Plot**    | `plt.scatter()`       | Displays individual data points.          |
| **Bar Plot**        | `plt.bar()`           | Shows categorical data as bars.           |
| **Histogram**       | `plt.hist()`          | Displays the frequency distribution.      |
| **Pie Chart**       | `plt.pie()`           | Represents proportions in a circle.       |
| **Box Plot**        | `plt.boxplot()`       | Displays data distribution using quartiles.|

# Line Chart

A **line chart** (or **line plot**) is a graphical representation of data points connected by straight lines. Line charts are most commonly used to display **trends** over time or the relationship between two variables. They are particularly useful when the data is continuous or when you want to highlight patterns in the data, such as growth, decline, or cyclical behavior.

---

### **Key Features of a Line Chart**

- **X-axis**: Typically represents the independent variable (e.g., time, categories).
- **Y-axis**: Represents the dependent variable (the value being measured).
- **Data Points**: Points plotted on the graph that represent the values of the variables.
- **Line**: A continuous line connecting the data points, showing the trend of the data.

---

### **Syntax of `plt.plot()`**
```python
plt.plot(x, y, linestyle='-', color='blue', marker='o', label='Line 1')
```

- **`x`**: The data for the x-axis (independent variable).
- **`y`**: The data for the y-axis (dependent variable).
- **`linestyle`**: Defines the style of the line (e.g., solid, dashed).
- **`color`**: Specifies the color of the line.
- **`marker`**: Adds markers to the data points.
- **`label`**: Adds a label to the line, used in the legend.

---

## Example: Multi-line Chart

````python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x+y for x, y in zip(variance, bias_squared)]
````

````python
x = [i for i, _ in enumerate(variance)]
````

> Using position and keyword arugments for `plot()`, **combining the color, marker, and linestyle**. 

````python
plt.plot(x, variance, 'gd--', label="variance")
plt.legend()
plt.show()
````

````python
plt.plot(x, variance, 'gd--', label="Variance")
plt.plot(x, bias_squared, 'bo-', label="Bias Sqaured")
plt.plot(x, total_error, 'rp-.', label="Total Error")
plt.legend()
plt.grid()
plt.show()
````

In the above the xticks don't mean anything, hence we can remove them. 

````python
plt.plot(x, variance, 'gd--', label="Variance")
plt.plot(x, bias_squared, 'bo-', label="Bias Sqaured")
plt.plot(x, total_error, 'rp-.', label="Total Error")
plt.legend()
plt.xlabel("Model Complexity")
plt.xticks([])
plt.title("The Bias-Variance Tradeoff")
plt.grid()
plt.show()
````

# Bar Charts

A **bar chart** is a graphical representation of categorical data where rectangular bars are used to represent the frequency or value of each category. The length of each bar is proportional to the value it represents. Bar charts are useful for comparing quantities across different categories.

In Matplotlib, bar charts are created using the `plt.bar()` function.

```python
plt.bar(x, height, color='color', width=0.8, label='label')
```
- **`x`**: Categorical data (e.g., labels).
- **`height`**: Values corresponding to each category.
- **`color`**: Color of the bars.
- **`width`**: Width of the bars.
- **`label`**: Label for the legend.

### When to use Bar Charts?
- To compare quantities across different categories.
- To visualize discrete data or categorical data.
- When you need to show the frequency or count of categories.
- To display data that belongs to distinct groups or categories.
- To compare multiple sets of categorical data (grouped bar charts).
- When the categories are independent of each other.
- To highlight differences or trends between categories.

````python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
````

````python
plt.bar(range(len(movies)), num_oscars)
plt.show()
````

````python
plt.bar(range(len(movies)), num_oscars)
plt.title("My Favorite Movies")
plt.ylabel("# of Academy Awards")

plt.xticks(range(len(movies)), movies)
plt.show()
````

#### Bar charts are also good for histograms of bucketed numeric values

````python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# Bucket grades by decile, but put 100 in with the 90s
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
````

````python
plt.bar([x + 5 for x in histogram.keys()], # Shift bars right by 5
        histogram.values(),                # Give each bar its correct height
        10,                                # Give each bar a width of 10
        edgecolor=(0, 0, 0))               # Black edges for each bar



plt.axis([-5, 105, 0, 5]) # x-axis from -5 to 105, y-axis from 0 to 5
plt.xticks([10 * i for i in range(11)]) # x-axis labels at 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
````

> Be judicious when using `plt.axis`. When creating bar charts it is considered especially bad form for your y-axis not to start at 0, since this is an easy way to mislead people

````python
mentions = [500, 505]
years = [2017, 2018]
````

````python
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

# if you don't do this, matplotlib will label the x-axis 0, 1
# and then add a +2.013e3 off in the corner (bad matplotlib!)
plt.ticklabel_format(useOffset=False)
# misleading y-axis only shows the part above 500
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Look at the 'Huge' Increase!")
plt.show()
````

````python

plt.axis([2016.5, 2018.5, 0, 550])
plt.title("Not So Huge Anymore")
plt.show()
````

# Scatter Plots

A **scatter plot** is a type of data visualization used to display the relationship between two continuous variables. It consists of points plotted on a two-dimensional grid, where each point represents a pair of values from the variables. Scatter plots are commonly used to identify patterns, trends, correlations, and outliers in the data.

In Matplotlib, scatter plots are created using the `plt.scatter()` function.

```python
plt.scatter(x, y, c='color', marker='o', label='label')
```
- **`x`**: Data for the x-axis.
- **`y`**: Data for the y-axis.
- **`c`**: Color of the points.
- **`marker`**: Style of the points (e.g., 'o', 's', etc.).
- **`label`**: Label for the legend.

---

### When to use Scatter Plots?
- To visualize the relationship between two continuous variables.
- To identify correlations (positive, negative, or no correlation).
- To detect outliers or anomalies in the data.
- To explore patterns or clusters in the data.
- To observe the distribution of data points across two dimensions.
- When you want to highlight individual data points or trends.
- To examine the spread or variance of data.

---

````python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)
````

#### Let's try to label each point.

````python
plt.scatter(friends, minutes)
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, 
                 xy=(friend_count, minute_count), # put the label with its point
                 xytext=(5,-5), # slight offset
                 textcoords='offset points')

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
````

### Important: 
Always ensure you choose the scale, and not to use matplotlib's default scale. 

````python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
````

#### To make Axes comparable:

````python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]
plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.axis("equal")
plt.show()
````
